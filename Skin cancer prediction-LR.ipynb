```python
data = pd.read_csv("54339_104884_bundle_archive/HAM10000_metadata.csv")
```

```python
data.sex = [1 if each == 'female' else 0 for each in data.sex]
```

```python
data.dx = [1 if each == 'bkl' or each == 'nv' or each == 'df' else 0 for each in data.dx]
```

```python
data['age'].dropna()
```

```python
data = data.dropna()
```

```python
new_categ = pd.get_dummies(data['localization'],prefix ='local')
```

```python
new_categ
```

```python
newdf = pd.concat([data,new_categ],axis=1)
```

```python
newdf.head()
```

```python
data = newdf
```

```python
data.head()
```

```python
new_categ_df = pd.get_dummies(data['dx_type'],prefix ='dx_type')
```

```python
new_data =  pd.concat([data,new_categ_df],axis=1)
```

```python
new_data.head()
```

```python
data = new_data
```

```python
data.head()
```

```python
data = data.drop(['localization','lesion_id','image_id','dx_type'],axis=1)
```

```python
data.head()
```

```python
data.tail()
```

```python

x = data.drop('dx',axis=1)
y = data['dx']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mylogmodel = LogisticRegression()
```

```python
mylogmodel.fit(x_train,y_train)
```

```python
skincancerpredictions = mylogmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,skincancerpredictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test,skincancerpredictions)
```

```python
accuracy
```

```python
cm = confusion_matrix(y_test, skincancerpredictions)
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```
